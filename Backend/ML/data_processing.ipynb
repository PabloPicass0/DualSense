```python
# Test Dataset and get_dataset
import numpy as np
from utils.dataset import Dataset
import matplotlib.pyplot as plt
from utils.dataset import Dataset

# Instantiate the Dataset object
dataset = Dataset(config_path='config_STSL.json')

# Display the first three images from X_train
for i in range(3):
    plt.imshow(dataset.X_train[i].squeeze(), cmap='gray')  # squeeze() is used to remove the singleton dimension if your images are grayscale
    plt.title(f"Label: {np.argmax(dataset.y_train[i])}")  # assuming y_train is one-hot encoded
    plt.show()

# Display the first image from X_test
plt.imshow(dataset.X_test[0].squeeze(), cmap='gray')
plt.title(f"Label: {np.argmax(dataset.y_test[0])}")  # assuming y_test is one-hot encoded
plt.show()

# print shape
print("X_train shape:", dataset.X_train.shape)
print("y_train shape:", dataset.y_train.shape)
print("X_test shape:", dataset.X_test.shape)
print("y_test shape:", dataset.y_test.shape)
```
